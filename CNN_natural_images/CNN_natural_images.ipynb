```python
from PIL import Image
from os import listdir
from numpy import asarray

names = ['airplane', 'car', 'cat', 'dog', 'flower', 'fruit', 'motorbike', 'person']

res_images_dict = {'airplane':[], 'car':[], 'cat':[], 'dog':[],
                   'flower':[], 'fruit':[], 'motorbike':[], 'person':[]}
```

```python
path = 'C://Users//Dhruvraj//Desktop//INEURON//10. DEEP LEARNING//natural_images'

for i in names:
    
    for image in listdir(path + '//' + i):
        img_data = Image.open(path + '//' + i +'//' +image)
        res_image = img_data.resize((150, 100))
        data = asarray(res_image)
        res_images_dict[i].append(data)
#         res_image.save('C://Users//Dhruvraj//Desktop//INEURON//10. DEEP LEARNING//Resized_Images//'+ image)
        print('> loaded %s' % (image), 'into', i)
    
```

```python
import numpy as np

names = ['airplane', 'car', 'cat', 'dog', 'flower', 'fruit', 'motorbike', 'person']
val = []

for i in names:
    total_images = np.array(res_images_dict[i]).shape[0]
    total = total + round(0.20*total_images, 0)
    print('{:10} => {:5} images, 20% => {} '.format(i, total_images, round(0.20*total_images, 0)))
    val.append((total_images, int(round(0.20*total_images, 0)) , int(total_images - round(0.20*total_images, 0))))
    
val
```

```python
info = dict(zip(names, val))
info
```

```python
# CREATING TRAIN AND VALIDATION DATASET

x1 = {}
x2 = {}
y1 = {}
y2 = {}
count = 0

for i, j in info.items():
    key1 = i+'_train'
    value1 = res_images_dict[i][:j[2]]
    
    key2 = i+'_test'
    value2 = res_images_dict[i][j[2]:]
    
    ans1 = count*np.ones((j[2]),dtype=int)
    ans2 = count*np.ones((j[1]),dtype=int)
    
    x1[key1] = value1
    x2[key2] = value2
    y1[key1] = ans1
    y2[key2] = ans2

    count += 1
```

```python
from kerastuner import RandomSearch
from kerastuner.engine.hyperparameters import HyperParameters
from keras.layers import LeakyReLU
from tensorflow import keras
from tensorflow.keras.layers import Input, Conv2D, Dense, Flatten, Dropout
from tensorflow.keras.models import Model

def build_model(hp):
    model = keras.Sequential([
        keras.layers.Conv2D(
            filters=hp.Int('conv_1_filter', min_value=32, max_value=128, step=16),
            kernel_size=hp.Choice('conv_1_kernel', values = [7]),
            activation='relu',
            input_shape=(100, 150, 3)
        ),
        keras.layers.Conv2D(
            filters=hp.Int('conv_2_filter', min_value=32, max_value=512, step=16),
            kernel_size=hp.Choice('conv_2_kernel', values = [5]),
            activation='relu'
        ),
        keras.layers.Conv2D(
            filters=hp.Int('conv_3_filter', min_value=32, max_value=512, step=16),
            kernel_size=hp.Choice('conv_3_kernel', values = [3]),
            activation='relu'
        ),
        keras.layers.Flatten(),
        keras.layers.Dense(
            units=hp.Int('dense_1_units', min_value=32, max_value=512, step=16),
            activation='relu'
        ),
        keras.layers.Dense(K, activation='softmax')
      ])

    model.compile(optimizer=keras.optimizers.Adam(hp.Choice('learning_rate', values=[1e-2, 1e-3])),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    return model
```

```python
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=3,
    executions_per_trial=4,
    directory= 'C://Users//Dhruvraj//Desktop//INEURON//10. DEEP LEARNING//',
    project_name='cnn_natural_images')
```
